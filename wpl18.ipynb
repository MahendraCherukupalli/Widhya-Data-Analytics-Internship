```python
strike_rate=data1[(data1["batting_team"].isin(['Royal Challengers Bangalore','Mumbai Indians']))&(data1["bowling_team"].isin(['Royal Challengers Bangalore','Mumbai Indians']))]
```

```python
df=pd.DataFrame()
```

```python
df["balls"]=strike_rate.groupby(["match_id","batsman"])["ball"].count()
```

```python
df["total_runs"]=strike_rate.groupby(["match_id","batsman"])["batsman_runs"].sum()
```

```python
strike_rate1=strike_rate[strike_rate["batsman_runs"].isin([4,6])]
```

```python
df["boundaries"]=strike_rate1.groupby(["match_id","batsman"])["batsman_runs"].count()
```

```python
df
```

```python
df=df.fillna(0)
```

```python
def status(x,y,z):
    if x>=10:
        if y>=3:
            strike=(z/x)*100
            if strike>=130:
                return "Yes"
            else:
                return "No"
        else:
            return "No"
    else:
        return "No"
```

```python
df["status"]=df.apply(lambda x: status(x.balls, x.boundaries,x.total_runs), axis=1)
```

```python
df
```

```python
df.groupby("match_id")["status"].value_counts()
```
