```python
strike_rate=data1[(data1["batting_team"].isin(['Delhi Daredevils',"Royal Challengers Bangalore","Delhi Capitals"]))&(data1["bowling_team"].isin(['Delhi Daredevils',"Royal Challengers Bangalore","Delhi Capitals"]))]

df=pd.DataFrame()

df["balls"]=strike_rate.groupby(["match_id","batsman"])["ball"].count()

df["total_runs"]=strike_rate.groupby(["match_id","batsman"])["batsman_runs"].sum()

strike_rate0=strike_rate[strike_rate["batsman_runs"].isin([4,6])]

df["boundaries"]=strike_rate0.groupby(["match_id","batsman"])["batsman_runs"].count()

strike_rate1=strike_rate[strike_rate["batsman_runs"]==4]

df["boundaries_4"]=strike_rate1.groupby(["match_id","batsman"])["batsman_runs"].count()

strike_rate2=strike_rate[strike_rate["batsman_runs"]==6]

df["boundaries_6"]=strike_rate2.groupby(["match_id","batsman"])["batsman_runs"].count()

df=df.fillna(0)

df
```

```python
def status(a,b,c,d):
    if a>=10:
        if b>=3:
            if c>=2:
                strike=(d/a)*100
                if strike>=130:
                    return "Yes"
                else:
                    return "No"
            else:
                return "No"
        else:
            return "No"
    else:
        return "No"
```

```python
df["status"]=df.apply(lambda x: status(x.balls, x.boundaries,x.boundaries_4,x.total_runs), axis=1)
```

```python
df.groupby("match_id")["status"].value_counts()
```

according to the dataset 3-4 batsmens satisfies the given condition

```python
economy_rate=data1[(data1["batting_team"].isin(['Delhi Daredevils',"Royal Challengers Bangalore","Delhi Capitals"]))&(data1["bowling_team"].isin(['Delhi Daredevils',"Royal Challengers Bangalore","Delhi Capitals"]))]
```

```python
df1=pd.DataFrame(economy_rate.groupby(["match_id","bowler"])["ball"].count())
```

```python
dot_balls1=economy_rate[economy_rate["total_runs"]==0]
```

```python
df1["dot_balls"]=dot_balls1.groupby(["match_id","bowler"])["ball"].count()
```

```python
economy_rate1=economy_rate[economy_rate["total_runs"].isin([4,6])]
df1["boundaries"]=economy_rate1.groupby(["match_id","bowler"])["total_runs"].count()
```

```python
df1["total_runs"]=economy_rate.groupby(["match_id","bowler"])["total_runs"].sum()
```

```python
df1["economy_rate"]=(df1["total_runs"])/(df1["ball"]/6)
```

```python
def status(a,b,c,d):
    if (a/6)>=2:
        if b>=8:
            if c<=4:
                if d<=8:
                    return "Yes"
                else:
                    return "NO"
                
            else:
                return "NO"
        else:
            return "No"
    else:
        return "No"
                
                
    
```

```python
df1["Status"]=df1.apply(lambda x: status(x.ball, x.dot_balls,x.boundaries,x.economy_rate), axis=1)
```
