```python
de_kock_BIS=data1[(data1["batting_team"].isin(['Rajasthan Royals','Mumbai Indians']))&(data1["bowling_team"].isin(['Rajasthan Royals','Mumbai Indians'])&(data1["batsman"]=="Q de Kock"))]
```

```python
df=pd.DataFrame()
```

```python
df["total runs"]=de_kock_BIS.groupby("match_id")["batsman_runs"].sum()
```

```python
de_kock_BIS1=de_kock_BIS[de_kock_BIS["batsman_runs"].isin([4,6])]
```

```python
de_kock_BIS2=de_kock_BIS[de_kock_BIS["batsman_runs"].isin([1,2,3])]
```

```python
df["boundary runs"]=de_kock_BIS1.groupby("match_id")["batsman_runs"].sum()
```

```python
df["123 runs"]=de_kock_BIS2.groupby("match_id")["batsman_runs"].sum()
```

```python
df["balls faced"]=de_kock_BIS.groupby("match_id")["ball"].count()
```

```python
df["BIS"]=((df["total runs"]+df["boundary runs"])-(df["123 runs"]*0.5))/df["balls faced"]
```

```python
archer=data1[(data1["batting_team"].isin(['Rajasthan Royals','Mumbai Indians']))&(data1["bowling_team"].isin(['Rajasthan Royals','Mumbai Indians'])&(data1["bowler"]=="J Archer"))]
df6=pd.DataFrame()
df6["total wickets"]=archer.groupby("match_id")["player_dismissed"].count()
archer1=archer[archer["batsman_runs"].isin([4,6])]
archer2=archer[archer["total_runs"]==0]
df6["boundary runs"]=archer1.groupby("match_id")["batsman_runs"].sum()
df6["dot balls"]=archer2.groupby("match_id")["batsman_runs"].count()
df6["balls bowld"]=archer.groupby("match_id")["ball"].count()
df6["BLS"]=((df6["total wickets"]+(df6["dot balls"]*2)-df6["boundary runs"]))/(df6["balls bowld"]-(df6["dot balls"]*2.5))
df6
```

```python
df6["BLS"].mean()
```
