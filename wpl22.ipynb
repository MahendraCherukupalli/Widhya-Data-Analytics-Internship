```python
stokes=data1[(data1["batting_team"].isin(['Rajasthan Royals','Kolkata Knight Riders']))&(data1["bowling_team"].isin(['Rajasthan Royals','Kolkata Knight Riders'])&(data1["batsman"]=="BA Stokes"))]
df1=pd.DataFrame()
df1["total runs"]=stokes.groupby("match_id")["batsman_runs"].sum()
stokes1=stokes[stokes["batsman_runs"].isin([4,6])]
stokes2=stokes[stokes["batsman_runs"].isin([1,2,3])]
df1["boundary runs"]=stokes1.groupby("match_id")["batsman_runs"].sum()
df1["123_runs"]=stokes2.groupby("match_id")["batsman_runs"].sum()
df1["balls_faced"]=stokes.groupby("match_id")["ball"].count()
df1["Ratio"]=(df1["boundary runs"]-df1["123_runs"])/(df1["total runs"])
df1
```

```python
df1["Ratio"].mean()
```

```python
archer=data1[(data1["batting_team"].isin(["Kolkata Knight Riders","Rajasthan Royals"]))&(data1["bowling_team"].isin(["Kolkata Knight Riders","Rajasthan Royals"])&(data1["bowler"]=="J Archer"))]
df=pd.DataFrame()
df["total_runs"]=archer.groupby("match_id")["total_runs"].sum()
df["balls_bowled"]=archer.groupby("match_id")["ball"].count()

archer1=archer[archer["batsman_runs"].isin([4,6])]
df["boundaries_given"]=archer1.groupby("match_id")["batsman_runs"].count()
df["economy_rate"]=(df["total_runs"])/(df["balls_bowled"]/6)
df
```
