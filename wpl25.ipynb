```python
ecomony_rate=data1[(data1["batting_team"].isin(["Mumbai Indians","Delhi Daredevils","Delhi Capitals"]))&(data1["bowling_team"].isin(["Mumbai Indians","Delhi Daredevils","Delhi Capitals"]))]

df=pd.DataFrame()

df["balls_bowled"]=ecomony_rate.groupby(["match_id","bowler"])["ball"].count()

df["given_runs"]=ecomony_rate.groupby(["match_id","bowler"])["total_runs"].sum()

ecomony_rate0=ecomony_rate[ecomony_rate["total_runs"].isin([4,6])]

df["boundaries"]=ecomony_rate0.groupby(["match_id","bowler"])["total_runs"].count()

ecomony_rate1=ecomony_rate[ecomony_rate["total_runs"]==4]

df["boundaries_4"]=ecomony_rate1.groupby(["match_id","bowler"])["total_runs"].count()

ecomony_rate2=ecomony_rate[ecomony_rate["total_runs"]==6]

df["boundaries_6"]=ecomony_rate2.groupby(["match_id","bowler"])["total_runs"].count()

df=df.fillna(0)

df
```

```python
def status(a,b,c,d):
    if a>7:
        if b<=6:
            if c==2:
                economy=(d/a)
                if economy<=9:
                    return "Yes"
                else:
                    return "No"
            else:
                return "No"
        else:
            return "No"
    else:
        return "No"
```

```python
df["status"]=df.apply(lambda x: status(x.balls_bowled, x.boundaries,x.boundaries_6,x.given_runs), axis=1)
```

```python
df
```

```python
df.groupby("match_id")["status"].value_counts()
```

According to the dataset 0-2 bowlers satisfied the given condition
