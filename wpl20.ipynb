```python
gayle_oop=data1[(data1["batting_team"].isin(['Rajasthan Royals','Kings XI Punjab']))&(data1["bowling_team"].isin(['Rajasthan Royals','Kings XI Punjab'])&(data1["batsman"]=="CH Gayle"))]
df11=pd.DataFrame()
df11["total runs"]=gayle_oop.groupby("match_id")["batsman_runs"].sum()
gayle_oop1=gayle_oop[gayle_oop["batsman_runs"].isin([4,6])]
gayle_oop2=gayle_oop[gayle_oop["batsman_runs"].isin([1,2,3])]
df11["boundary runs"]=gayle_oop1.groupby("match_id")["batsman_runs"].sum()
df11["123_runs"]=gayle_oop2.groupby("match_id")["batsman_runs"].sum()
df11["balls faced"]=gayle_oop.groupby("match_id")["ball"].count()
df11["OOP"]=(df11["total runs"]/df11["balls faced"])+(df11["boundary runs"])-(df11["123_runs"]/df11["balls faced"])
df11
```

```python
df11["OOP"].mean()
```
