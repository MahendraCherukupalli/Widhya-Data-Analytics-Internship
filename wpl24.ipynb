```python
warner=data1[(data1["batting_team"].isin(['Sunrisers Hyderabad','Mumbai Indians']))&(data1["bowling_team"].isin(['Sunrisers Hyderabad','Mumbai Indians'])&(data1["batsman"]=="DA Warner"))]
crease=data1[(data1["batting_team"].isin(['Sunrisers Hyderabad','Mumbai Indians']))&(data1["bowling_team"].isin(['Sunrisers Hyderabad','Mumbai Indians'])&(data1["non_striker"]=="DA Warner"))]
df1=pd.DataFrame()
df1["total runs"]=warner.groupby("match_id")["batsman_runs"].sum()
warner1=warner[warner["batsman_runs"].isin([4,6])]
warner2=warner[warner["batsman_runs"].isin([1,2,3])]
df1["boundary runs"]=warner1.groupby("match_id")["batsman_runs"].sum()
df1["123_runs"]=warner2.groupby("match_id")["batsman_runs"].sum()
df1["balls faced"]=warner.groupby("match_id")["ball"].count()
df1["balls_in_crease"]=crease.groupby("match_id")["ball"].count()
df1["balls_in_crease"]=df1["balls faced"]+df1["balls_in_crease"]
df1["Stamina"]=((df1["123_runs"]/df1["boundary runs"])+(df1["balls faced"]/df1["balls_in_crease"]))/df1["total runs"]
df1
```
