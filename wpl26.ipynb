```python
strike_rate=data1[(data1["batting_team"].isin(['Royal Challengers Bangalore',"Sunrisers Hyderabad"]))&(data1["bowling_team"].isin(['Royal Challengers Bangalore',"Sunrisers Hyderabad"]))]

df1=pd.DataFrame()

df1["balls"]=strike_rate.groupby(["match_id","batsman"])["ball"].count()

df1["total_runs"]=strike_rate.groupby(["match_id","batsman"])["batsman_runs"].sum()

strike_rate0=strike_rate[strike_rate["batsman_runs"].isin([4,6])]

df1["boundaries"]=strike_rate0.groupby(["match_id","batsman"])["batsman_runs"].count()

strike_rate1=strike_rate[strike_rate["batsman_runs"]==4]

df1["boundaries_4"]=strike_rate1.groupby(["match_id","batsman"])["batsman_runs"].count()

strike_rate2=strike_rate[strike_rate["batsman_runs"]==6]

df1["boundaries_6"]=strike_rate2.groupby(["match_id","batsman"])["batsman_runs"].count()

df1=df1.fillna(0)

df1
```

```python
def status(a,b,c,d):
    if a>=10:
        if b>=3:
            if c>=2:
                strike=(d/a)*100
                if strike>=130:
                    return "Yes"
                else:
                    return "No"
            else:
                return "No"
        else:
            return "No"
    else:
        return "No"
```

```python
df1["status"]=df1.apply(lambda x: status(x.balls, x.boundaries,x.boundaries_4,x.total_runs), axis=1)
```

```python
df1.groupby("match_id")["status"].value_counts()
```

according to the dataset 3-4 batsmens satisfies the given condition

```python
kohli=data1[(data1["batting_team"].isin(['Royal Challengers Bangalore',"Sunrisers Hyderabad"]))&(data1["bowling_team"].isin(['Royal Challengers Bangalore',"Sunrisers Hyderabad"])&(data1["batsman"]=="V Kohli"))]
df2=pd.DataFrame()
df2["total runs"]=kohli.groupby("match_id")["batsman_runs"].sum()
kohli1=kohli[kohli["batsman_runs"].isin([4,6])]
kohli2=kohli[kohli["batsman_runs"].isin([1,2,3])]
df2["boundary runs"]=kohli1.groupby("match_id")["batsman_runs"].sum()
df2["123_runs"]=kohli2.groupby("match_id")["batsman_runs"].sum()
df2["balls_faced"]=kohli.groupby("match_id")["ball"].count()
df2["Ratio"]=(df2["boundary runs"]-df2["123_runs"])/(df2["total runs"])
df2
```
